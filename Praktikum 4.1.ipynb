```python
import matplotlib.pyplot as plt
%matplotlib inline ##untuk menampilkan gambar tanpa perlu menampilkan window baru
import numpy as np
import cv2

from skimage import data
from skimage.io import imread
from skimage.color import rgb2gray
```

```python
citra1 = imread(fname="noisyimage.jpg")
print(citra1.shape)
plt.imshow(citra1,cmap='gray')
```

```python
kernel = np.array([[-1, 0, -1],
                  [0, 4.3, 0],
                  [-1, 0, -1]])
citraoutput = cv2.filter2D(citra1, -1, kernel)

fig, axes = plt.subplots(1, 2, figsize=(12, 12))
ax = axes.ravel()

ax[0].imshow(citra1, cmap = 'gray')
ax[0].set_title("Citra Input")
ax[1].imshow(citraoutput, cmap = 'gray')
ax[1].set_title('Citra Output')
```

```python
img = cv2.imread("danautoba.jpg")
```

```python
cv2.imshow("ini danau", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
print(img.shape)
```

```python
[baris, kolom] = img.shape[:2]
```

# Membuat Histogram

```python
hist = cv2.calcHist([img], [0], None,[256], [0,256])
hist
```

```python
fig, axs =  plt.subplots(1,2, figsize = (15, 5))
axs[0].imshow(img)
axs[1].plot(hist, color = 'purple')
plt.show()
```

# Convert BGR to RGB

```python
rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
fig, axs =  plt.subplots(1,2, figsize = (15, 5))
axs[0].imshow(rgb)
axs[1].plot(hist, color = 'red')
plt.show()
```

# 2 Versi Histogram

```python
fig, axs =  plt.subplots(2,2, figsize = (22, 7))
# histogram versi 1
axs[0,0].imshow(img)
axs[0,1].plot(hist, color = 'purple')

# histogram versi 2
axs[1,0].imshow(img)
axs[1,1].hist(img.ravel(), 256, [0,256], color = 'purple')
plt.show()
```

# Histogram Abu-abu

```python
abu = cv2.imread('danautoba.jpg', 0)
hist = cv2.equalizeHist(abu)

fig, axs = plt.subplots(1,2, figsize = (15, 5))
axs[0].imshow(cv2.cvtColor(abu,cv2.COLOR_BGR2RGB))
axs[1].hist(abu.ravel(), 256, [0,256], color = "purple")
plt.show()
```

# Meningkatkan Kecerahan

```python
beta = 33
citra_cerah = np.zeros((baris, kolom, 3))

for x in range(baris):
    for y in range (kolom):
        gyx = img[x,y]+beta
        citra_cerah[x,y] = gyx
citra_cerah = citra_cerah.astype(np.uint8)

# sebelum
fig, axs = plt.subplots(2,2, figsize = (21, 7))
axs[0,0].imshow(img)
axs[0,1].hist(img.ravel(), 256, [0,256], color = "purple")

# sesudah
axs[1,0].imshow(citra_cerah)
axs[1,1].hist(citra_cerah.ravel(), 256, [0,256], color = "g")
plt.show()
```

# Meregangkan Kontras

```python
alpa = 33
citra_cerah = np.zeros((baris, kolom, 3))

for x in range(baris):
    for y in range (kolom):
        gyx = img[x,y]+alpa
        citra_cerah[x,y] = gyx
citra_cerah = citra_cerah.astype(np.uint8)

# sebelum
fig, axs = plt.subplots(2,2, figsize = (21, 7))
axs[0,0].imshow(img)
axs[0,1].hist(img.ravel(), 256, [0,256], color = "purple")

# sesudah
axs[1,0].imshow(citra_cerah)
axs[1,1].hist(citra_cerah.ravel(), 256, [0,256], color = "g")
plt.show()
```
