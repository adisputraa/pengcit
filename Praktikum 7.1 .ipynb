### Import Library

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

### Deteksi Gambar & Tepi

```python
img = cv.imread('nama_b.jpg')
```

```python
cv.imshow('Nama Saya', img)
cv.waitKey(0)
cv.destroyAllWindows()
```

### Menampilkan Tepi pada Gambar

```python
gray =  cv.cvtColor(img, cv.COLOR_BGR2GRAY)
edges = cv.Canny(img, 100, 150) # edges digunakan ketika akan mendeteksi pada sebuah citra
# canny merupakan sebuah fungsi untuk mendeteksi sebuah tepi
# 100, 150 merupakan jarak parameter, semakin besar angka yang dimasukkan maka tepi gambar akan semakin tipis
```

### Tampilkan Menggunakan Figure dan Axis (Fungsi matplotlib)

```python
# Menggunakan matplotlib
fig, axs = plt.subplots(1,2, figsize = (10, 10))
ax = axs.ravel()

# Tampilan Asli
ax[0].imshow(gray, cmap = 'gray')
ax[0].set_title('Citra Asli')

# Tampilan Edges
ax[1].imshow(edges, cmap = 'gray')
ax[1].set_title('After Edges')
```

### Menampilkan Tepi pada Gambar

```python
lines = cv.HoughLinesP(edges, 1, np.pi/90, 30, maxLineGap = 250) 
# cv.HoughLinesP digunakan untuk mendeteksi garis, 
# max untuk menentukan ambang batas gambarnya
# np.pi/180 dignuakan sebagai parameter dari teta untuk resolusinya
img_line = img.copy()
```

```python
for line in lines:
    x1, y1, x2, y2, = line[0]
    # x1, y1 merupakan sudut kiri atas garis
    # x2, y2 merupakan sudut kanan bawah garis
    cv.line(img_line, (x1,y1), (x2,y2), (0,0,225), 1)
    # (0,0,225) merupakan pemilihan warna
    # ,1 merupakan ketebalan garis pada tepi
```

### Tampilan Menggunakan Figure & Axis (Fungsi Matplotlib)

```python
gray1 = cv.cvtColor(img_line, cv.COLOR_BGR2RGB)

# Menggunakan matplotlib
fig, axs = plt.subplots(1,2, figsize = (10, 10))
ax = axs.ravel()

# Tampilan Asli
ax[0].imshow(gray, cmap = 'gray')
ax[0].set_title('Citra Asli')

# Tampilan Edges
ax[1].imshow(gray1, cmap = 'gray')
ax[1].set_title('Setelah Edges')
```
